##  Jordy DF for Word Cloud

```python
def word_cloud(comp_name):
    data =df.loc[df["company"] == comp_name]
    pros = []
    cons = []
    for index, row in data.iterrows():
        pros.append(row["pros"])
        cons.append(row["cons"])

    pros_str = ' '.join(pros)
    gptext = str(pros_str)

    # Create the wordcloud object
    gpwordcloud = WordCloud(width=1000, height=880, margin=0, max_font_size=100, max_words=50, background_color="white").generate(gptext)

    # Display the generated image:
    pwc = plt.figure(1)
    plt.imshow(gpwordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.margins(x=0, y=0)
    plt.title(f"{comp_name} Pro's")



    #Google Word Cloud - CON'S
    cons_str = ' '.join(cons)
    gctext = str(cons_str)

    # Create the wordcloud object
    gcwordcloud = WordCloud(width=1000, height=880, margin=0, max_font_size=100, max_words=50, background_color="white").generate(gctext)

    # Display the generated image:
    cwc = plt.figure(2)
    plt.imshow(gcwordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.margins(x=0, y=0)
    plt.title(f"{comp_name} Con's")

    plt.show()
```
