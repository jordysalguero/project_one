##  Jordy DF for Word Cloud

```python
def word_cloud(comp_name):
    data =df.loc[df["company"] == comp_name]
    pros = []
    cons = []
    for index, row in data.iterrows():
        pros.append(row["pros"])
        cons.append(row["cons"])

    pros_str = ' '.join(pros)
    pros_text = str(pros_str)

    # Create the wordcloud object
    pro_wordcloud = WordCloud(width=1000, height=880, margin=0, max_font_size=100, max_words=50, background_color="white").generate(pros_text)

    # Display the generated image:
    pwc = plt.figure(1)
    plt.imshow(pro_wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.margins(x=0, y=0)
    plt.title(comp_name.capitalize() + " " +  "Pro's")
    plt.savefig("./Images/Word_Cloud/" + comp_name.capitalize() + "_" + "Pro's" + ".png")



    #Company Word Cloud - CON'S
    cons_str = ' '.join(cons)
    cons_text = str(cons_str)

    # Create the wordcloud object
    cons_wordcloud = WordCloud(width=1000, height=880, margin=0, max_font_size=100, max_words=50, background_color="white").generate(cons_text)

    # Display the generated image:
    cwc = plt.figure(2)
    plt.imshow(cons_wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.margins(x=0, y=0)
    plt.title(comp_name.capitalize() + " " +  "Con's")
    plt.savefig("./Images/Word_Cloud/" + comp_name.capitalize() + "_" + "Con's" + ".png")

    plt.show()
```
