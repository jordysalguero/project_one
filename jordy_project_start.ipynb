```python
import numpy as np
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import pandas as pd
import matplotlib.pyplot as plt
from function_creation import draw_chart,draw_chart2, print_label, calc_means, calc_sems, count_sample
```

```python
#read, select, and rename the columns
df = pd.read_csv("employee_reviews.csv",encoding="utf-8")
#df=df[['Unnamed: 0','company', 'location','dates','job-title','overall-ratings','work-balance-stars','culture-values-stars','carrer-opportunities-stars','comp-benefit-stars','senior-mangemnet-stars']]
#df=df.rename(columns={'Unnamed: 0':'0','job-title':'title','overall-ratings':'overall', 'work-balance-stars':'wk_bal','culture-values-stars':'culture_val','carrer-opportunities-stars':'career_opp','comp-benefit-stars':'comp_benefit', 'senior-mangemnet-stars':'Sr_mgmt'})
df.head()
```

##  Jordy DF for Word Cloud

```python
#Google Word Cloud - PRO's
df_google =df.loc[df["company"] == "google"]
google_pros = []
google_cons = []
for index, row in df_google.iterrows():
    google_pros.append(row["pros"])
    google_cons.append(row["cons"])
    
google_pros_str = ' '.join(google_pros)
gptext = str(google_pros_str)

# Create the wordcloud object
gpwordcloud = WordCloud(width=1000, height=880, margin=0, max_font_size=100, max_words=50, background_color="white").generate(gptext)

# Display the generated image:
gpwc = plt.figure(1)
plt.imshow(gpwordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.title("Google Pro's")



#Google Word Cloud - CON'S
google_cons_str = ' '.join(google_cons)
gctext = str(google_cons_str)

# Create the wordcloud object
gcwordcloud = WordCloud(width=1000, height=880, margin=0, max_font_size=100, max_words=50, background_color="white").generate(gctext)

# Display the generated image:
gcwc = plt.figure(2)
plt.imshow(gcwordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.title("Google Con's")

plt.show()
```
